# To prove the statement::
### y = μ + ε

### where y = observed value
### μ = mean expected value from the observation
### ε = error term
#### The error term ε is represented by the symbol N(0,1) and has a mean of 0 and a standard deviation of 1.

### Here we need to find the mean μ that minimizes the sum of squared errors (SSE). The SSE is defined as:
# SSE = Σ(ε^2)

### ε is defined as the deviation of the observed value from the mean (ε = y - μ). The condition for a minimum is met by taking the derivative with respect to μ and setting it equal to zero in order to minimise the SSE.

### Differentiating :
### d(SSE)/dμ = d/dμ Σ((y - μ)^2) = 0

### Expanding inside the summation:
### d/dμ Σ((y^2 - 2yμ + μ^2)) = 0

### Simplifying further:
### Σ(-2y + 2μ) = 0

### Multiplying out the summation:
### -2Σy + 2Nμ = 0

### Separating μ on one side and dividing by 2N gives us the formula for the mean μ that minimizes the SSE:
### μ = (Σy) / N

### Hence, proved


# Problem 2(b)

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating samples
np.random.seed(0)  
mu_true = 1
epsilon = np.random.randn(1000)  # 1000 samples from N(0, 1)
y_samples = mu_true + epsilon

# Estimating mu by minimizing SSE
mu_candidates = np.arange(0, 2.01, 0.01)
SSE = []

for mu in mu_candidates:
    sse = np.sum((y_samples - mu) ** 2)
    SSE.append(sse)

# SPlot SSE vs mu_candidates
plt.figure(figsize=(10, 5))
plt.plot(mu_candidates, SSE, label='SSE')
plt.xlabel('mu candidates')
plt.ylabel('SSE')
plt.title('SSE vs. mu Candidates')
plt.legend()
plt.grid(True)

# Annotate the minimum point on the plot
min_sse_index = np.argmin(SSE)
min_sse_mu = mu_candidates[min_sse_index]
min_sse_value = SSE[min_sse_index]
plt.annotate(f'min SSE at mu = {min_sse_mu}', xy=(min_sse_mu, min_sse_value), 
             xytext=(min_sse_mu+0.5, min_sse_value+200), arrowprops=dict(arrowstyle='->', lw=1.5))

plt.show()

# Return the mu that minimizes SSE
print("The value of mu that minimizes the SSE is ")
min_sse_mu
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data
data = {
    'consumers': [1, 1.08, 1.15, 1.15, 1.2, 1.3, 1.37, 1.37, 1.43, 1.46,
                  1.52, 1.57, 1.65, 1.65, 1.65, 1.66, 1.87, 2.03, 2.05, 2.3],
    'acres': [1.71, 1.52, 1.29, 3.09, 2.21, 2.26, 2.4, 2.1, 1.96, 2.09,
              2.02, 1.31, 2.17, 2.28, 2.41, 2.23, 3.04, 2.06, 2.73, 2.36]
}

# Create DataFrame
df = pd.DataFrame(data)

# Divide data into three groups
group1 = df.head(7)
group2 = df.iloc[7:13]
group3 = df.tail(7)

# Calculate mean X and mean Y for each group
mean_group1 = group1.mean()
mean_group2 = group2.mean()
mean_group3 = group3.mean()

# Plot scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='consumers', y='acres', data=df, color='blue', alpha=0.7)
plt.title('Scatterplot of Acres/Gardener vs Consumers/Gardener')

# Add mean points
plt.scatter(mean_group1['consumers'], mean_group1['acres'], color='red', s=100, label='Group 1 Mean')
plt.scatter(mean_group2['consumers'], mean_group2['acres'], color='green', s=100, label='Group 2 Mean')
plt.scatter(mean_group3['consumers'], mean_group3['acres'], color='orange', s=100, label='Group 3 Mean')

# Connect mean points with nonparametric regression line
plt.plot([mean_group1['consumers'], mean_group2['consumers'], mean_group3['consumers']],
         [mean_group1['acres'], mean_group2['acres'], mean_group3['acres']],
         linestyle='-', color='black')

plt.xlabel('Consumers/Gardener')
plt.ylabel('Acres/Gardener')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

### The scatterplot now includes the mean values for all three home groups, connected by a simple nonparametric regression line (in red). This line graphically illustrates the relationship between 'Acres/Gardener' and 'Consumers/Gardener' among the three groups.The regression line indicates a positive association between the two variables. However, the line may not adequately represent the true underlying relationship because it is based on the mean values of groups that have been rather arbitrarily divided.
