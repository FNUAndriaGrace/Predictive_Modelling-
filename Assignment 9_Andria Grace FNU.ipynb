```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.arima_process import ArmaProcess

# Model (a) parameters
phi_a = 0.6
theta_a = 0.4

# Simulate ARMA(1,1) process for model (a)
ar_params = np.array([1, -phi_a])
ma_params = np.array([1, theta_a])
arma_process_a = ArmaProcess(ar_params, ma_params)
sample_size = 100  # Number of samples

# Generate ARMA(1,1) time series data
np.random.seed(0)
data_a = arma_process_a.generate_sample(nsample=sample_size)

# Plot autocorrelations for model (a)
acf_a = sm.tsa.acf(data_a, nlags=20)
plt.figure(figsize=(8, 6))
plt.stem(acf_a, use_line_collection=True)
plt.title("Autocorrelations of ARMA(1,1) Process (φ=0.6, θ=0.4)")
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Define parameters for each ARMA(1,1) model
models = [
    {'phi': 0.9, 'theta': 0.2},    # Model (a)
    {'phi': -0.9, 'theta': 0.2},   # Model (b)
    {'phi': 0.9, 'theta': -0.2},   # Model (c)
    {'phi': -0.9, 'theta': -0.2}   # Model (d)
]

# Function to simulate ARMA(1,1) process and plot ACF of errors
def simulate_arma_and_plot_acf(phi, theta, model_label):
    # Simulate ARMA(1,1) process
    np.random.seed(0)
    n = 100  # Number of time points
    ar_params = np.r_[1, -phi]
    ma_params = np.r_[1, theta]
    arma_process = sm.tsa.ArmaProcess(ar_params, ma_params)
    simulated_data = arma_process.generate_sample(nsample=n)

    # Fit ARMA(1,1) model to the simulated data to obtain errors
    model = sm.tsa.AutoReg(simulated_data, lags=1).fit()
    errors = model.resid

    # Plot ACF of errors
    plt.figure(figsize=(8, 4))
    sm.graphics.tsa.plot_acf(errors, lags=30, alpha=0.05, title=f"ACF of Errors - {model_label}")
    plt.show()

# Simulate and plot each ARMA(1,1) model
for i, model_params in enumerate(models):
    phi = model_params['phi']
    theta = model_params['theta']
    model_label = f"Model ({chr(97 + i)})"  # Generate label (a), (b), (c), (d)
    simulate_arma_and_plot_acf(phi, theta, model_label)
```

Showing Equivalence between Recursive EWMA and Infinite Sum MA(1)
To show that the recursive Exponentially Weighted Moving Average (EWMA) calculation produces the value as the infinite sum of MA(1), we need to derive the relationship and show the equivalence between the two.

Given the recursive EWMA equation:
𝑚𝑡 = 𝜃𝑚𝑡−1 + 𝜖𝑡

 We can rewrite this recursively as:
mt = ϵt + θmt−1
   = ϵt +θ(ϵt−1 +θmt−2 )
   =ϵt +θϵt−1 +θ^2mt−2
mt =ϵt + θϵt−1 +θ(ϵt−2)…

In the infinite sum form, this becomes:
𝑚𝑡 = ∑∞ 𝑘=0 𝜃^𝑘 𝜖𝑡−𝑘

This shows that the recursive EWMA calculation can be represented as an infinite sum of the MA(1) process, where each term in the sum is multiplied by the corresponding power of the parameter  𝜃.
