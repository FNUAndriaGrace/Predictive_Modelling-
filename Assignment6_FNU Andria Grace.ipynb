## 1a) Write out lnL(B) for logistic regression. Omit the constant term and use one regressor so that the model has only one parameter, B.

The log-likelihood function ln L(B) for logistic regression with one regressor and one parameter B:

Let, 
Yi denote the binary response variable for the ith observation,
Xi denote the value of the regressor for the ith observation.

The logistic regression model assumes:

P(Yi = 1∣Xi) = 1 / (1 + e^−(BXi)

P(Yi = 0∣Xi) = 1 − 1 + e^−(Bxi)

The likelihood function for a single observation i:

Li(B) = Yiln (1 / (1 + e^−(Bxi)) + (1- Yi)ln (1 − (1 / 1 + e^−(Bxi))

Taking the natural logarithm of the likelihood function, we get:

lnLi(B) =Yiln (1 / (1+ e^−(Bxi)) + (1 − Yi)ln (1 − (1 / 1 + e^−(Bxi))

The log-likelihood function for all observations can be obtained by summing over all observations:

lnL(B) = n ∑ i=1 lnLi(B)

Where n is the total number of observations.

## 1b) Also write it using NumPy in a Python function like "log_likelihood(x,y,B)".


```python
import numpy as np

def log_likelihood(x, y, B):
    
    # Calculating the logistic function
    logistic_func = 1 / (1 + np.exp(-B * x))
    
    # Calculating the likelihood for each observation
    log_likelihoods = y * np.log(logistic_func) + (1 - y) * np.log(1 - logistic_func)
    
    # Summing over all observations to get the total log-likelihood
    total_log_likelihood = np.sum(log_likelihoods)
    
    return total_log_likelihood
```
